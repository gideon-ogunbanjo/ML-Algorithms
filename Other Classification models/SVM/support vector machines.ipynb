Using SVM to Classfiy Different types of Iris Species

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#loading the dataset
data = pd.read_csv("Iris.csv")
```

```python
#returns info axis
data.keys()
```

```python
#return sample row
data.sample()
```

```python
#shape of the dataset(extensively)
data.describe()
```

```python
#shape of the dataset
data.shape
```

```python
#loading onto X and Y values
x = data.iloc[:,:-1] #iloc is used to select rows and colums in the order that they appear in the dataset
y = data.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
#convert categorical into numerical value
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
#transforming y values
y = lb.fit_transform(y)
```

```python
#spliting the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
#shape of the dataset
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#linear SVM model
svc_clf = SVC(kernel = 'linear')
svc_clf.fit(x_train, y_train)
y_pred = svc_clf.predict(x_test)
```

```python
#confusion matrix
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
#checking the accuracy and classification report
print(accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
#rbf SVM model
svc_clf_rbf = SVC(kernel='rbf')
svc_clf_rbf.fit(x_train, y_train)
y_pred = svc_clf_rbf.predict(x_test)
```

```python
#confusion matrix
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
#checking the accuracy and classification report
print(accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))
```
